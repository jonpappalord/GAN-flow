```python
with open("./"+ transp + city+ "/v_train.txt", "rb") as fp:   # Unpickling
    v_train = pickle.load(fp)
print(len(v_train))


with open("./"+ transp + city+ "/v_test.txt", "rb") as fp:   # Unpickling
    v_test = pickle.load(fp)
print(len(v_test))
```

```python
f_max = np.max(v_train)
print(f_max)
```

```python
a = 0
fake_set = []
f_max = np.max(v_train)
 
for n in sample(v_train, len(v_test)):
    randomlist = []
    for i in range(0,(64*64)):
        r = random.randint(0,f_max)
        randomlist.append(r)
    r_a = np.array(randomlist).reshape(64,64)
    fake_set.append(r_a)
```
