```python
def plot_JS_hist(exp1, exp2, exp3, ax, measure, metric, ylabel = False):    
    kwargs = dict(histtype='stepfilled', ec="k",  facecolor='none', density = True)
    
    min_v = min (exp1+exp2+exp3)
    max_v = max (exp1+exp2+exp3)
    
    r = np.linspace(min_v,max_v,20)
    k, bins = np.histogram(r,20)
    
    #bins = 20
    sns.kdeplot(exp1, label = "Test", bw_method=0.5, ax=ax, color = "red",   linewidth=1)
    sns.kdeplot(exp2, label = "Synthetic", bw_method=0.5, ax=ax, color = "blue",  linewidth=1)
    sns.kdeplot(exp3, label = "Mixed", bw_method=0.5, ax=ax, color = "green", linewidth=3)
    
    #INTER-DISTRIBUTION js
    values_3, base_3 = np.histogram(exp3, bins=bins, density=1)
    values_1, base_1 = np.histogram(exp1, bins=bins, density=1)
    js1 = distance.jensenshannon(np.asarray(values_3), np.asarray(values_1), np.e)
    
    values_2, base_2 = np.histogram(exp2, bins=bins, density=1)
    js2 = distance.jensenshannon(np.asarray(values_2), np.asarray(values_1), np.e)

    #text1 = str("JS(m,t)=%.2f" % js1)
    #text2 = str("JS(s,t)=%.2f" % js2)
    #text = text1 + '\n' + text2
    
    #ax.annotate(text, xy=(0.451, 0.99), xycoords='axes fraction', fontsize = 32, style = 'italic')    
    
    ax.tick_params(labelsize=34)
    ax.set_xlabel(measure, fontsize = 34)
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    if ylabel:
        ax.set_ylabel("density", fontsize = 34)
    else:
        ax.set_ylabel("", fontsize = 34)

    
    return (js1, js2)
    #ax.legend( prop={'size': 18})
```
